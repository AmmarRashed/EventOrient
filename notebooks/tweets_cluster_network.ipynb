```python
from AdaptiveOnlineClustering import AdaptiveOnlineClustering as AOC
import gensim, pickle, json, networkx as nx, numpy as np
```

```python
model = gensim.models.KeyedVectors.load_word2vec_format('/home/ammar/NLP_data/GoogleNews-vectors-negative300.bin', binary=True)  
```

```python
aoc = AOC(model, None)
```

```python
tweets = pickle.load(open('../datasets/tweets.pkl', 'rb'))
```

```python
import time
start = time.time()
i = 0
for author in tweets:
    for tweet in author:
        try:
            aoc.add(tweet.text, tweet.lang, tweet.author.id)
        except Exception as e:
            print(e)
            continue
        i+=1
        print("Added {} tweets".format(i))
print("Took {} seconds".format(time.time()-start))
```

```python
print("There are {} clusters of {} documents".format(len(aoc.clusters), i))
```

```python
graph = aoc.get_network_graph(print_prob=0.01)
```

```python
nx.write_gpickle(graph,"../datasets/tweets_cluster_graph_sample.gpickle")
```

```python
graph = nx.read_gpickle("../datasets/tweets_cluster_graph_sample.gpickle")
```

```python
evc = nx.eigenvector_centrality(graph)
closeness = nx.closeness_centrality(graph)
betweenness = nx.betweenness_centrality(graph)
```

```python
metrics = {"eigenvector_centrality":evc,
           "closeness_centrality":closeness,
          "betweenness":betweenness}
```

```python
for metric_name, metric in metrics.items():
    for ix,v in metric.items():
        graph.node[ix][metric_name] = v
for ix,deg in list(graph.degree()):
    graph.node[ix]['degree'] = deg
    graph.node[ix]['parity'] = (1-deg%2)
```

```python
authors_dict = pickle.load(open("../datasets/tweets_id2auth.pkl", "rb"))
```

```python
sample_author = authors_dict[765555429640470528]
sample_author
```

```python
augment= {"name":lambda x: x.name,
          "followers_count":lambda x: x.followers_count,
          "following_count":lambda x: x.friends_count,
          "language":lambda x:x.lang}
```

```python
for author_id in graph.nodes():
    author = authors_dict[author_id]
    for aug, func in augment.items():
        graph.nodes[author_id][aug] = func(author)
```

```python
from networkx.readwrite import json_graph
graph_data = nx.node_link_data(graph)
with open('../datasets/twitter_cluster_graph.json', 'w') as f:
    json.dump(graph_data, f, indent=4)
```

```python
mini_graph = nx.Graph()
portion = 0.01
for fr, to in graph.edges():
    if np.random.choice(a=[False, True], p=[1-portion, portion]):
        mini_graph.add_edge(fr,to, weight=graph.edges[(fr,to)]['weight'])
```
