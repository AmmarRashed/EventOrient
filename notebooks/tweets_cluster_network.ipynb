```python
from joblib import Parallel, delayed
```

```python
def tovec(tweet):
    return (tweet.text,aoc.vectorize(tweet.text, tweet.lang))
```

```python
data = dict()
authors = dict()
i = 0
for u in tweets:
    try:
        author = u[0].author
        vecs_ = Parallel(verbose=10, n_jobs=-1)(delayed(tovec)(tweet)
                                                      for tweet in u)
        vecs = {t:v for t,v in vecs_ if v is not None}
        data[author.id] = vecs
        authors[author.id] = author
    except Exception as e:
        print(e)
        continue
pickle.dump(data, open("../datasets/tweets_auth2tweets2vecs.pkl", "wb"), protocol=2)
pickle.dump(authors, open("../datasets/authid2auth.pkl", "wb"), protocol=2)
```

```python
import copy
def fill_nodes(graph, authors_dict, augment_dict):
    for author_id in graph.nodes():
        author = authors_dict[author_id]
        for aug, func in augment_dict.items():
            graph.nodes[author_id][aug] = func(author)
    return graph

def add_metrics(graph, metrics_):
    metrics = copy.deepcopy(metrics_)
    for metric, func in metrics.items():
        metrics[metric] = func(graph)
    for metric_name, metric in metrics.items():
        for ix,v in metric.items():
            graph.node[ix][metric_name] = v
    for ix,deg in list(graph.degree()):
        graph.node[ix]['degree'] = deg
        graph.node[ix]['parity'] = (1-deg%2)
    return graph

augment= {"name":lambda x: x.name,
          "twitter_name":lambda x: x.screen_name,
          "followers_count":lambda x: x.followers_count,
          "following_count":lambda x: x.friends_count,
          "language":lambda x:x.lang}

metrics = {"eigenvector_centrality":nx.eigenvector_centrality,
           "closeness_centrality":nx.closeness_centrality,
          "betweenness":nx.betweenness_centrality}
```

```python
graph = fill_nodes(graph, authors, augment)
```

```python
graph = add_metrics(graph, metrics)
```

```python
from networkx.readwrite import json_graph
graph_data = nx.node_link_data(graph)
with open('../visualization/twitter_cluster_graph.json', 'w') as f:
    json.dump(graph_data, f, indent=4)
# with open('../datasets/twitter_cluster_graph.json', 'w') as f:
#     json.dump(graph_data, f, indent=4)
```

```python
def get_mini_graph(graph, portion=0.01, authors=None, augment=None, metrics=None, weighted=True):
    mini_graph = nx.Graph()
    for fr, to in graph.edges():
        if np.random.choice(a=[False, True], p=[1-portion, portion]):
            if weighted:
                mini_graph.add_edge(fr,to, weight=graph.edges[(fr,to)]['weight'])
            else:
                mini_graph.add_edge(fr,to)
    if len(authors) and len(augment):
        mini_graph = fill_nodes(mini_graph, authors, augment)
    if metrics:
        mini_graph = add_metrics(mini_graph, metrics)
    return mini_graph
```

```python
mini_graph = get_mini_graph(graph, 0.01, authors, augment, metrics)
```
